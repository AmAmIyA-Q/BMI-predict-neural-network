```python
import numpy as np
import os
from tqdm import tqdm

I2 = []

data_dir = 'data_mixed/I2'
files = [f for f in os.listdir(data_dir) if f.endswith('.txt')]

for f in tqdm(files):
    I2.append(np.loadtxt(data_dir + '/' + f))
```

```python
I2 = np.array(I2)
np.mean((I2-0.05)**2)
```

```python
# 直方图
import matplotlib.pyplot as plt

plt.hist(I2, bins=50)
```
